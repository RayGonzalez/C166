```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class IsingModel:
    
    def __init__(self, n = 10, T = 1):
        self.n = n
        self.T = T
        self.grid = np.zeros(shape = (n,n))
        self.random_start()
    
    
    def show(self):
        plt.matshow(self.grid)
        plt.show()
            
    
    def random_start(self):
        for i in range(self.n):
            for j in range(self.n):
                self.grid[i,j] = np.random.choice([-1,1])
                
    
    def update(self, steps = 10000):
        for _ in range(steps):
        
            i, j = np.random.randint(0, self.n, 2)

            E = -self.grid[i, j] * (self.grid[(i+1)%self.n, j] + self.grid[(i-1)%self.n, j] 
                                    + self.grid[i, (j+1)%self.n] + self.grid[i, (j-1)%self.n])

            if np.random.random() < np.exp(2*E/self.T):
                self.grid[i, j] *= -1
                
    def magnetization(self):
        return 1.0/(self.n**2) * self.grid.sum()
    
```

### Example start

```python
surface = IsingModel()
surface.show()
```

### T = 1

```python
surface = IsingModel()
surface.update()
surface.show()
```

```python
mag_list = []
for i in range(100):
    surface = IsingModel()
    surface.update()
    mag_list.append(surface.magnetization())
plt.hist(mag_list)
plt.title("Magnetization Frequencies, T = 1")
plt.show()
```
